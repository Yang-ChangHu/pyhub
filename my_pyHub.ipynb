输入矩阵，画混淆矩阵图

```python
import sklearn
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import itertools
import pylab
pylab.rcParams['figure.figsize'] = (13.0, 8.0)

matplotlib.rcParams['font.family'] = 'SimHei'

def plot_confusion_matrix(cm
                          ,classes
                          ,normalize=False
                         ,title='subject1:confusion matirx without normalizing'
                         ,cmap=plt.cm.Blues
                         ):
    #cm :混淆矩阵
    #classes ：类别 列表
    if normalize:
        cm=cm.astype('float')/cm.sum(axis=1)[:,np.newaxis]
        print('Normalize confusion matirx')
        
    else:
        print('confusion matirx without nomalization')
    print(cm)
    
    plt.imshow(cm,interpolation='nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks=np.arange(len(classes))
    plt.xticks(tick_marks,classes,rotation=45)
    plt.yticks(tick_marks,classes)
    
    fmt='.2f' if normalize else 'd'
    thresh=cm.max()/2.
    for i,j in itertools.product(range(cm.shape[0]),range(cm.shape[1])):
        plt.text(j
                ,i
                ,format(cm[i,j],fmt)
                ,horizontalalignment='center'
                ,color='white' if cm[i,j]>thresh else 'black')
        plt.ylabel('True label')
        plt.xlabel('predicted label')
        plt.tight_layout()
        

cm_total=pd.read_csv('C:\\Users\\ych\\Desktop\\lectures\\GMM\\data\\subject1_6.csv')
```
